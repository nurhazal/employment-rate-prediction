```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/istihdam-verileri.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['istihdam_orani'].describe()
```

```python
# Kirilim Bir (bar plot)
plt.figure(figsize=(10,4))
sns.countplot(data=df, x='kirilim_bir')
plt.title('Kirilim Bir Kategorileri Sayısı')
plt.xlabel('Kirilim Bir')
plt.ylabel('Sayı')
plt.tight_layout()
plt.savefig("../outputs/raw/kirilim_bir_countplot.png")
plt.close()

# Kirilim Iki (bar plot)
plt.figure(figsize=(12,4))
sns.countplot(data=df, x='kirilim_iki')
plt.title('Kirilim İki Kategorileri Sayısı')
plt.xlabel('Kirilim İki')
plt.ylabel('Sayı')
plt.tight_layout()
plt.savefig("../outputs/raw/kirilim_iki_countplot.png")
plt.close()

# İstihdam Oranı – Histogram + KDE
plt.figure(figsize=(8,4))
sns.histplot(df['istihdam_orani'], bins=20, kde=True)
plt.title('İstihdam Oranı Dağılımı')
plt.xlabel('İstihdam Oranı (%)')
plt.ylabel('Frekans')
plt.tight_layout()
plt.savefig("../outputs/raw/istihdam_orani_distribution.png")
plt.close()

# İstihdam Bin Kişi vs İstihdam Oranı – Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(
    x=df['istihdam_bin_kisi'],
    y=df['istihdam_orani']
)
plt.title('İstihdam Bin Kişi vs İstihdam Oranı')
plt.xlabel('İstihdam Bin Kişi')
plt.ylabel('İstihdam Oranı (%)')
plt.tight_layout()
plt.savefig("../outputs/raw/istihdam_bin_vs_oran.png")
plt.close()

# Yıl bazında istihdam oranlarının dağılımı (Boxplot)
plt.figure(figsize=(10,5))
sns.boxplot(x=df['yil'], y=df['istihdam_orani'])
plt.title('Yıllara Göre İstihdam Oranı Dağılımı')
plt.xlabel('Yıl')
plt.ylabel('İstihdam Oranı (%)')
plt.tight_layout()
plt.savefig("../outputs/raw/yillara_gore_istihdam_orani.png")
plt.close()
```
